```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Normalization method
# Minmaxscaler  shrinks the data within the range of -1 to 1(if there are negative values)
# This responds well if standard deviation is small and  is used when distribution is not Gaussian.
# This scaler is sensitive to outliers.
def MinMaxScaler(data):
    numerator = data - np.min(data, 0)
    denominator = np.max(data, 0) - np.min(data, 0)
    return numerator / (denominator + 1e-7)
#     return numerator / (denominator)

# train parameters
seq_length = 50        # sequence length
input_dim = 1          # dimension of input value
output_dim = 1         # dimenstion of output value
learning_rate = 0.01
iterations = 100


# Step 1. Data preporation
# Load dataset
df = pd.read_csv('eurofxref-hist.csv')

# 5 years data for test
df = df[df['Date'] >= '2014-12-16']

# US dollar for test
df = df.loc[:, ['USD']]
df = df[::-1].values
xy = np.array(df)

# train/test split
train_size = int(len(xy) * 0.9)
train_set = xy[0:train_size]
test_set = xy[train_size - seq_length:]

# Normalization
# The goal of normalization is to change the values of numeric columns in the dataset to use a common scale,
# without distorting differences in the ranges of values or losing information
train_set = MinMaxScaler(train_set)
test_set = MinMaxScaler(test_set)


# Step 2. Dataset seperation

# build the dataset method
def build_window(time_series, seq_length):
    dataX = []
    dataY = []
    for i in range(0, len(time_series) - seq_length):
        _x = time_series[i:i + seq_length, :]
        _y = time_series[i + seq_length, [-1]]
        dataX.append(_x)
        dataY.append(_y)
    return np.array(dataX), np.array(dataY)

# Dividing training dataset and testing dataset
X_train, y_train = build_window(train_set, seq_length)
X_test, y_test = build_window(test_set, seq_length)


# Step 3. Modelling

# Building model (layer by layer)
tf.model = tf.keras.Sequential()

# Selecting LSTM model
tf.model.add(tf.keras.layers.LSTM(units=1, input_shape=(seq_length, input_dim)))

# Dense layer is just a regular layer of neurons in a neural network.
tf.model.add(tf.keras.layers.Dense(units=output_dim, activation='tanh'))

# Summary of model
tf.model.summary()

# Tuning model
tf.model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(lr=learning_rate), metrics=['accuracy'])


# Step 4. Traning

# Training model
tf.model.fit(X_train, y_train, epochs=iterations)


# Step 5. Testing

# Testing model
y_predict = tf.model.predict(X_test)


# Step 6. Visualization

# Visualization model
plt.style.use('seaborn')
plt.figure(figsize=(20, 10))
plt.xticks(size=10, rotation='horizontal')
plt.plot(y_test, marker='o')
plt.plot(y_predict, marker='o')
plt.title('Currency Exchange Rates', size=40)
plt.xlabel("Period", size=20)
plt.ylabel("Rates", size=20)
plt.legend(labels=['Acutal Rates', 'Prediction Rates' ], loc='best', fontsize=20)
plt.show()
```
